```python
import numpy as np
import pandas as pd
```

```python
# Read data from file

df = pd.read_csv("https://raw.githubusercontent.com/freeCodeCamp/boilerplate-demographic-data-analyzer/master/adult.data.csv")
df.head()
```

```python
df.info()
```

```python
# How many of each race are represented in this dataset? This should be a Pandas series with race names as the index labels.
race_count = df["race"].value_counts()
race_count
```

```python
# What is the average age of men?
men = df["sex"] == "Male"
average_age_men =df.loc[(men), "age"].mean()
average_age_men
```

```python
# What is the percentage of people who have a Bachelor's degree?
bac = len(df[df["education"] == "Bachelors"])
education = len(df["education"])
percentage_bachelors = 100 * (bac/education)
percentage_bachelors
```

```python
# with and without `Bachelors`, `Masters`, or `Doctorate`

highdegree_list = ['Bachelors','Masters','Doctorate']  

                   
higher_education = df[df['education'].isin(highdegree_list)]
lower_education = df[~df['education'].isin(highdegree_list)]

higher_education

```

```python
# percentage with salary >50K


higher_education_rich = 100 * (higher_education.salary ==">50K").value_counts(normalize=True)[True]
lower_education_rich = 100 * (lower_education.salary ==">50K").value_counts(normalize=True)[True]

lower_education_rich
```

```python
# What is the minimum number of hours a person works per week (hours-per-week feature)?
min_work_hours = df["hours-per-week"].min()
min_work_hours
```

```python
# What percentage of the people who work the minimum number of hours per week have a salary of >50K?
num_min_workers = df.loc[(df["hours-per-week"] == 1) & (df["salary"] == ">50K")]
min_work = df.loc[(df["hours-per-week"] == 1)]

rich_percentage = 100 * len(num_min_workers) / len(min_work)
rich_percentage
```

```python
# What country has the highest percentage of people that earn >50K?
rich_count = df.loc[(df["salary"] == ">50K")]["native-country"].value_counts()
count = df["native-country"].value_counts()

richest = (rich_count / count).max()

                       
highest_earning_country = (rich_count/count).sort_values(ascending = False).index[0]
highest_earning_country
```

```python
highest_earning_country_percentage = 100 * richest
highest_earning_country_percentage
```

```python
# Identify the most popular occupation for those who earn >50K in India.
india = df.loc[(df["salary"] == ">50K") & ( df["native-country"] == "India")]["occupation"]
top_IN_occupation = pd.Series.mode(india)[0]
top_IN_occupation
```
